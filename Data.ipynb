```python
import pandas as pd
import urllib
import json
import csv
```

```python
poi_reviews=pd.read_csv("C:\\Users\\dhane\\Documents\\NLP Project\\Data\\rome-poi.csv")
acc_reviews=pd.read_csv("C:\\Users\\dhane\\Documents\\NLP Project\\Data\\barcelona-accommodation.csv")
res_reviews=pd.read_csv("C:\\Users\\dhane\\Documents\\NLP Project\\Data\\paris-restaurant.csv")
attr_reviews=pd.read_csv("C:\\Users\\dhane\\Documents\\NLP Project\\Data\\amsterdam-attraction.csv",error_bad_lines=False)
```

```python
review_links=pd.concat([poi_reviews,acc_reviews,res_reviews,attr_reviews])
```

```python
review_links.info()
review_links=review_links.dropna()
```

```python
review_links=review_links['reviews'].tolist()
```

```python
import requests
count = 0
data_file = open('data_file.csv', 'w', encoding="utf-8") 
csv_writer = csv.writer(data_file) 
for i,url in enumerate(review_links):
    r = requests.get(url)   
    for rev in r.json(): 
        if count == 0:   
            # Writing headers of CSV file 
            header = rev.keys() 
            csv_writer.writerow(header) 
            count += 1  
        # Writing data of CSV file 
        csv_writer.writerow(rev.values())   
data_file.close()
```
